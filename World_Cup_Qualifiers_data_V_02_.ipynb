#**Importing libraries**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

##**General Scraping Function**




```python
def scrap(link, your_columns):

  # --------- Get the the whole page --------- #

  response = requests.get(str(link))
  data = response.text
  soup = BeautifulSoup(data,"html.parser")

  # --------- Get the the desired data from the page --------- #

  Teams_stats = soup.find_all('tr')

  teams_data = [] # collecting the data in a list 

  for team in Teams_stats:  

    team_data = []
    Team_stats = team.find_all("td")  

    for team in Team_stats:

      team_data.append(team.text) 

    teams_data.append(team_data)

  # --------- Remove empty lists --------- #

  teams_data = [list_ for list_ in teams_data if list_ != []]

  # --------- Remove extra letters in the names --------- #

  for team in teams_data:
    team[0] = team[0][3:]

  # --------- Convert to a dataframe --------- #

  teams_data = pd.DataFrame(teams_data, columns = your_columns )  # inserting columns names

  return teams_data
```

#**Scraping Africa Data**

```python
Africa = scrap("https://fbref.com/en/comps/2/WCQ----CAF-M-Stats", ['Squad',	'MP',	'W',	'D',	'L',	'GF',	'GA',	'GD', 'Pts', 'Notes'])
Africa
```
